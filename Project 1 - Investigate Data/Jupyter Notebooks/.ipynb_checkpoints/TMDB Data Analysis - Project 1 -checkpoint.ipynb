### Removing duplicates

> Only one movie is repeated twice

```python
# drop the duplicated rows
df.drop_duplicates(ignore_index=True, inplace=True)
```

```python
## show the columns with NAs data
df.isna().sum() 
```

> There are 23 N/A genres, I need to check if those missing data will affect our analysis.

```python
# introducing a new column which is the profit of the movie.
df['profit'] = df['revenue_adj'] - df['budget_adj']
```

```python
# data with non-english characters
# Get all the columns that have non-ascii characters, that may affect the analysis
non_ascii_columns = []
for col in df.select_dtypes(object):
    if df[col].str.contains(r'[^\x00-\x7F]+', regex=True).any(axis=0):
        non_ascii_columns.append(col)
non_ascii_columns   
```

> The above columns have data with Non-English characters.

### Denormalizing the data, and breaking the columns with multiple values

```python
## this function to break the row that contains composite column into multiple rows 
def denormalize_data (df, col, sep=r'\|' ):
    if df[col].str.contains(sep).any():
        df['temp'] = df[col].str.split(sep) # a temp column is used to avoid changing the main dataframe to avoid changing the granularity
        denormalized_df =  df.explode('temp', ignore_index=True)
        df.drop(columns='temp', inplace = True)
        denormalized_df[col] = denormalized_df['temp'] 
        denormalized_df.drop(columns='temp', inplace = True)
        return denormalized_df
```

```python
# Denormalize & split the multivalue columns

genre_df= denormalize_data( df,'genres')
companies_df = denormalize_data(df,'production_companies')
```

> The main answer to our questions are about the genres and companies, so we need to break the genres, and production companies columns into new columns with only one value per row.

```python
# Check the unique information count in the newly generated dataframe 
genre_df.loc[:, ['genres']].nunique()
```

```python
companies_df['production_companies'].unique().size
```

```python
# Count the number of n/a generes
genre_df['genres'].isna().sum()
```

> After replicating the movies based on the genres they have, there are still 23 movies without a genre.

```python
# Checking the impact of the N/A genres on the profit analysis 
genre_df[genre_df['genres'].isna()].profit.sum()/genre_df.profit.sum() * 100
```

> The proportion of the profit of the missing genres to the total profit of the dataset is alomst zero, so I can remove those data.

```python
# eleminating the movies with no genres
genre_df.dropna(inplace=True, subset=['genres'])

# now check again
genre_df['genres'].isna().sum()
```

# Exploratory Data Analysis


## Questions
## Q1) What were the most profitable movie genres through the past 55 years?


```python
# Aggregating the dataset by the release year and movie genre.

genre_summary = genre_df.loc[:  , ['release_year','genres','vote_count', 'vote_average', 'profit', 'popularity', 'imdb_id']].groupby(as_index= False, by=['release_year','genres']).agg({'vote_count': 'sum', 'vote_average': 'mean', 'profit': 'sum', 'popularity':'mean', 'imdb_id':  pd.Series.nunique}).sort_values(by=['release_year','profit'], ascending=[True, False], ignore_index=True).rename(columns={'imdb_id':'movies_count'})

# Making sure we have no duplicates
genre_summary.loc[:, ['release_year', 'genres']].duplicated().sum()
```

#### Showing Top Genres (Sorted by Profit)

```python
genre_summary
```

```python
df[(df['release_year'] == 2015) & (df['genres'].str.contains('Mystery', na=False)) ]
```

#### Showing the top Geners over all the period

```python
genre_summary.groupby('genres')['profit'].sum().sort_values( ascending=False).apply( lambda x: '${}B'.format(round(x/1000000000, 1))).to_frame()
```

```python

# Getting the top 3 genres per the measure we analyze
def get_top_genres (measure):
    return genre_summary.groupby('genres', as_index=False)[measure].sum().sort_values(by=measure, ascending=False, ignore_index= True)['genres'][0:3].tolist()


# showing trendlines for the different measures we analyze
def trend_lines (measure, genres_list=None):
    color= {'C0', 'C1', 'C2', 'C3', 'C4', 'C5'}
    top_genres  = get_top_genres(measure) if genres_list is None else genres_list 
    df = genre_summary[genre_summary['genres'].str.contains ('|'.join(top_genres))]
    all_genres= df['genres'].unique().tolist()
   
    fig, ax = plt.subplots(figsize=(15,8))
    for i in all_genres:
        x = df[df['genres'].str.contains(i)] ['release_year'].tolist()
        y = df[df['genres'].str.contains(i)][measure].tolist()
        plt.scatter(x, y)
        plt.plot(x,y, label= i, linewidth=3)
    
    plt.title("The {} of top genres over years".format( measure))
    plt.ylabel("Measure")
    plt.xlabel("Release Years")
    plt.legend()
    plt.grid()
    plt.xticks(rotation = 0)
    plt.show()
```

```python
trend_lines('profit')
trend_lines('movies_count', get_top_genres('profit'))
```

> As shown in the above charts, although *Drama* movies are produced more than the *Adventure* movies almost each year, the adventure and action movies are more profitable.

```python
trend_lines('vote_average', get_top_genres('profit'))
trend_lines('vote_count', get_top_genres('profit'))
```

> AS per the above graphs, although the *Adventure* and *Action* movies the most profitable, people give them less rating than the *Drama* movies.

### Displaying The Top 10 Profitable Companies Over The Years 

```python
companies_df.groupby(by='production_companies')['profit'].sum().sort_values(ascending=False).apply( lambda x: '${}B'.format(round(x/1000000000, 1))).to_frame()[0:10]
```

```python
# Aggregating the data by the production company and release year.

companies_agg = companies_df.loc[:  , ['production_companies', 'release_year', 'profit', 'popularity', 'imdb_id']].groupby(as_index= False, by=['release_year','production_companies']).agg({'profit': 'sum', 'popularity':'mean', 'imdb_id':  pd.Series.nunique}).sort_values(by=['profit'], ascending= False, ignore_index=True).rename(columns={'imdb_id':'movies_count'})


# Then summarizing the data by only the production company
companies_summary = companies_agg.groupby('production_companies', as_index=False).agg({'profit': 'mean', 'popularity': 'mean', 'movies_count': 'mean'}).sort_values(by='profit', ascending=False)
```

```python
companies_summary
```

```python
## show a bar chart 

def show_bar_chart(df, measure):
    # order by the measure in ascending order 
    new_df = df.sort_values(by=measure, ascending= False)[0: 10].iloc[::-1] # get only the top 10 
    index = new_df['production_companies'].values
    y = new_df[measure].values

    # create plot
    fig, ax = plt.subplots(figsize=(15,8))
    bar_width = 0.65
    opacity = 0.75

    chart1 = plt.barh(index, y, bar_width,
                     alpha=opacity,
                     color='b')

    plt.xlabel(measure)
    plt.ylabel('Company Name')
    count = len(new_df)
    plt.title('Top {} Companies By Mean {} Per Year'.format(count, measure))

    plt.tight_layout()
    plt.show()
```

```python
show_bar_chart(companies_summary, 'movies_count')
```

```python
show_bar_chart(companies_summary, 'profit')
```

> As per the above graphs, the companies produce higher number of movies per year are not necessarily the most profitable yearly. 
> *Werner Bros* as instance is the 3rd higher movies producer but it is on the 10th bar when it comes to the mean profit per year.
